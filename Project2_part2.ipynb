```python
#(1)(1)IDENTIFY THE MOST COMMON COMBINATION OF CUISINES IN THE DATASET.

common_cusine_combi = data['Cuisines'].value_counts()
common_cusine_combi
```

```python
#(2)DETERMINE IF CERTAIN CUISINES COMBINATION TEND TO HAVE HIGHER RATINGS.

high_rat = data.groupby('Cuisines')['Aggregate rating'].mean().sort_values()
high_rat.tail(20)

```
